### Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[5,7,4])
plt.show()
```

### Legends, Titles, and Labels

```python
x1=[1,2,3]
y1=[5,7,4]

x2=[1,2,3]
y2=[10,14,12]
```

```python
#Here we are doing multiple plotting
plt.plot(x1,y1,label='First Line')
plt.plot(x2,y2,label='Second Line')
plt.xlabel('Plot Number')                         #Giving the x label
plt.ylabel('Important var')                       #Giving the y label
plt.title('Interesting Graph\nCheck it out')      #Giving the title for this graph
plt.legend()                                      #Invoking default legend. Legend will not show if we don't give label in plot()
plt.show()
```
