```python
import matplotlib.pyplot as plt
plt.bar([1,3,5,7,9],[5,2,7,8,2],label="Bar 1")
plt.bar([2,4,6,8,10],[8,6,2,5,6], label="Bar 2", color='g')
plt.xlabel('Bar Number')
plt.ylabel('Bar Height')
plt.legend()
plt.title("Bar Graph in Action")
plt.show()
```
