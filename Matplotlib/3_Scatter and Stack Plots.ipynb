## Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]

plt.scatter(x,y, label='skitscat', color='k', s=25, marker="o")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()
plt.show()

#The plt.scatter allows us to not only plot on x and y, but it also lets us decide on the color, size, and type of marker we use
#For more marker options see :: https://matplotlib.org/api/markers_api.html
```

## Stack Plot

```python
#Here we consider situation of 24hrs in a day. We havve divided the whole day in 4 categories
#Sleeping, Eating, Working, Playing
#Our x-axis will be days

days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating =   [2,3,4,3,2]
working =  [7,8,7,2,2]
playing =  [8,5,7,8,13]
```

```python
plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.show()
```
