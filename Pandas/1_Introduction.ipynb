```python
#Importing necessary files
import pandas as pd
import datetime
from pandas_datareader import data
```

```python
#Setting the start date and end date for which we want the data
start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2016, 8, 22)
```

```python
#Pulling data from Internet
#This pulls data for Exxon from the Yahoo Finance API, storing the data to our df variable
df = data.DataReader("XOM", "yahoo", start, end)
```

```python
#Printing first five records
print(df.head())
```

```python
#Since we want to visualize our data we will use matplotlib
import matplotlib.pyplot as plt
from matplotlib import style
style.use('fivethirtyeight')
```

```python
#Generating the graph
df['High'].plot()
plt.legend()
plt.show()
```
